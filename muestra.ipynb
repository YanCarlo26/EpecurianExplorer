```python
import json 
import glob 
import pandas as pd
import matplotlib.pyplot as plt
```

**Sabor Habanero: Explorando las Zonas y Sabores con Flow**

La Habana, con su mezcla única de historia, cultura y ritmo, no solo es un deleite para los sentidos, sino también para el paladar. Esta guía te invita a un viaje gastronómico por los barrios más vibrantes y diversos de la ciudad. Desde las arenas doradas de Habana del Este hasta los acogedores rincones de Boyeros y la animada vida urbana de Centro Habana, cada zona ofrece una experiencia culinaria que refleja la esencia misma de Cuba.

En esta aventura, exploraremos la distribución de restaurantes y bares, descubriremos los mejores horarios para disfrutar de una buena comida, y nos adentraremos en la variedad de tipos de comida que La Habana tiene para ofrecer. También analizaremos los métodos de pago, la presencia en redes sociales, los servicios que hacen única cada experiencia, y compararemos los rangos de precios entre estas fascinantes zonas.

Prepárate para una inmersión completa en el sabor habanero,, con todos los detalles y análisis que te ayudarán a saborear La Habana al máximo. ¡Vamos a darle flow a este recorrido gastronómico!

```python
#Cargando los json
def open_json():
    nombres = []
    zone = []
    Location = []
    place_type = []
    food_type =[]
    schedule =[]
    menu = []
    
    #Busca en el directorio "./Jsons/*" los municipios 
    for municipio in glob.glob("./Jsons/*"):
        #Busca dentro de los municipios los archivos que sean .json 
        for rest in glob.glob(municipio + "/*.json"):
            #Abre y lee cada uno de los archivos .json
            with open(rest, "r", encoding="utf-8") as f:
                data = json.load(f)
                nombres.append(data["Name"])
                zone.append(data["Zone"])
                Location.append(data["Location"])
                place_type.append(data["Place Type"])
                food_type.append(data["Food Type"])
                schedule.append(data["Schedule"])
                menu.append (data["Menu"])
                
    
    return nombres, zone, Location, place_type, food_type, schedule, menu
        


nombres, zone, Location, place_type, food_type, schedule, menu,  = open_json()               
display(nombres)
display(zone)
display(Location)
display(place_type)
display(food_type)
display(schedule)
display(pd.DataFrame (menu))
```


Imagina un mapa de La Habana dividido en coloridas secciones, cada una representando una zona distinta. Los restaurantes y bares están diseminados por toda la ciudad, con algunas zonas destacándose por su mayor concentración de lugares para comer y beber. A través de un gráfico de pastel, observamos cómo estas zonas se dividen equitativamente o se concentran en ciertas áreas, permitiéndonos entender mejor la distribución de la oferta culinaria en la ciudad.

Por ejemplo, el famoso barrio del Vedado es conocido por su vida nocturna y su gran cantidad de bares y restaurantes. Desde elegantes establecimientos hasta acogedores locales familiares, esta zona ofrece una experiencia culinaria para todos los gustos. Por otro lado, la Habana Vieja, con sus calles adoquinadas y su arquitectura colonial, también alberga una considerable cantidad de restaurantes que combinan el encanto histórico con una oferta gastronómica variada.

Este análisis no solo destaca la variedad disponible, sino que también permite a los visitantes planificar sus recorridos gastronómicos de manera eficiente, asegurando que siempre haya algo nuevo y emocionante por descubrir en cada zona de la ciudad.

```python
# Análisis de distribución geográfica
zone_counts = {}
for z in zone:
    if z in zone_counts:
        zone_counts[z] += 1
    else:
        zone_counts[z] = 1


# Crear gráfico de pastel
plt.figure(figsize=(8, 8))
plt.pie(zone_counts.values(), labels=zone_counts.keys(), autopct='%1.1f%%')
plt.title('Distribución de Restaurantes por Zona')
plt.show()
```


La diversidad culinaria de La Habana es asombrosa. Desde comida asiática hasta cocina tradicional cubana, cada restaurante ofrece una experiencia única para los amantes de la gastronomía. El análisis de los tipos de comida revela la amplitud de opciones disponibles, permitiendo a los comensales explorar nuevos sabores y satisfacer sus antojos.

Por ejemplo, los amantes de la comida asiática pueden disfrutar de sushi fresco y platos exóticos en restaurantes como Asia Cuba, mientras que aquellos que prefieren la cocina italiana pueden deleitarse con pastas hechas a mano y pizzas gourmet. Además, la cocina tradicional cubana, con sus platos de arroz con pollo, ropa vieja y tostones, ofrece una experiencia auténtica que transporta a los visitantes a las raíces culturales de la isla.

Este análisis es crucial para los foodies que buscan descubrir cada rincón de la ciudad a través de sus platos y disfrutar de una experiencia culinaria diversa y enriquecedora.

```python
# Análisis de tipos de comida
food_type_counts = {}
for food in food_type:
    if isinstance(food, list):
        print(f"Error: {food} es una lista en lugar de un string.")
    else:
        if food in food_type_counts:
            food_type_counts[food] += 1
        else:
            food_type_counts[food] = 1
# Crear gráfico de barras
plt.figure(figsize=(10, 6))
plt.bar(food_type_counts.keys(), food_type_counts.values())
plt.title('Tipos de Comida Ofrecidos')
plt.xlabel('Tipo de Comida')
plt.ylabel('Número de Restaurantes')
plt.xticks(rotation=45)
plt.show()
```


La Habana es una ciudad que nunca duerme, y sus restaurantes y bares reflejan este espíritu. A través del análisis de los horarios de apertura y cierre, descubrimos patrones interesantes sobre cuándo los establecimientos están más activos. Los restaurantes que abren temprano suelen ofrecer deliciosos desayunos y brunches que deleitan tanto a los madrugadores como a aquellos que buscan comenzar su día con una experiencia culinaria excepcional.

Por otro lado, aquellos que prefieren la cena y la vida nocturna encontrarán una amplia variedad de bares y restaurantes que permanecen abiertos hasta altas horas de la noche, ofreciendo no solo comidas, sino también entretenimiento en vivo, cócteles exquisitos y un ambiente animado. Este análisis te permitirá encontrar el lugar perfecto según tu horario, asegurando que siempre tengas una opción disponible sin importar la hora del día

```python

# Diccionarios para contar las frecuencias de apertura y cierre
open_times = {}
close_times = {}

for sched in schedule:
    for day_info in sched:
        if 'open' in day_info and day_info['open']:
            if day_info['open'] in open_times:
                open_times[day_info['open']] += 1
            else:
                open_times[day_info['open']] = 1
        if 'close' in day_info and day_info['close']:
            if day_info['close'] in close_times:
                close_times[day_info['close']] += 1
            else:
                close_times[day_info['close']] = 1

# Convertir las claves a cadenas si es necesario
x_labels_open = list(map(str, open_times.keys()))
x_labels_close = list(map(str, close_times.keys()))

# Crear gráfico de barras para horarios de apertura
plt.figure(figsize=(10, 6))
plt.bar(x_labels_open, open_times.values(), color='blue')
plt.title('Horarios de Apertura')
plt.xlabel('Hora de Apertura')
plt.ylabel('Frecuencia')
plt.show()

# Crear gráfico de barras para horarios de cierre
plt.figure(figsize=(10, 6))
plt.bar(x_labels_close, close_times.values(), color='green')
plt.title('Horarios de Cierre')
plt.xlabel('Hora de Cierre')
plt.ylabel('Frecuencia')
plt.show()


```


En una era donde la conveniencia es clave, los métodos de pago disponibles en los restaurantes juegan un papel importante. Nuestro análisis muestra cuáles son las opciones más comunes, desde pagos en efectivo hasta pagos en línea. Con esta información, los visitantes pueden planificar mejor sus visitas, asegurándose de que siempre tendrán una forma de pagar que les resulte conveniente.

Por ejemplo, muchos restaurantes modernos en La Habana aceptan pagos en línea y tarjetas de crédito, lo que facilita la experiencia para los turistas internacionales. Al mismo tiempo, los locales más tradicionales pueden preferir el efectivo, especialmente en moneda local (CUP). Este análisis ayuda a los visitantes a estar preparados y a evitar sorpresas desagradables al momento de pagar.

```python
# Análisis de métodos de pago
payment_options = []
for municipio in glob.glob("./Jsons/*"):
    for rest in glob.glob(municipio + "/*.json"):
        with open(rest, "r", encoding="utf-8") as f:
            data = json.load(f)
            payment_options.append(data["Payment Options"])

payment_counts = {}
for options in payment_options:
    for method, accepted in options.items():
        if isinstance(accepted, dict):  # Para pagos en efectivo
            for currency, is_accepted in accepted.items():
                method_currency = f"{method} ({currency})"
                if is_accepted:
                    if method_currency in payment_counts:
                        payment_counts[method_currency] += 1
                    else:
                        payment_counts[method_currency] = 1
        else:
            if accepted:
                if method in payment_counts:
                    payment_counts[method] += 1
                else:
                    payment_counts[method] = 1

# Crear gráfico de barras
plt.figure(figsize=(10, 6))
plt.bar(payment_counts.keys(), payment_counts.values())
plt.title('Métodos de Pago Aceptados')
plt.xlabel('Método de Pago')
plt.ylabel('Número de Restaurantes')
plt.xticks(rotation=45)
plt.show()
```


Las redes sociales han revolucionado la forma en que descubrimos nuevos lugares. La presencia en redes sociales de los restaurantes de La Habana no solo les permite alcanzar a más clientes, sino que también ofrece una ventana al mundo para que los visitantes puedan ver reseñas, menús y fotos antes de decidir a dónde ir.

Plataformas como Instagram y Facebook son especialmente populares entre los restaurantes que desean mostrar sus platos más fotogénicos y atraer a un público más joven y conectado digitalmente. Las publicaciones y las reseñas de los clientes proporcionan información valiosa sobre la calidad de la comida, el servicio y el ambiente de cada lugar. Este análisis destaca qué tan activos están los restaurantes en estas plataformas y cómo esto influye en su popularidad.

```python
# Análisis de presencia en redes sociales
social_media_presence = []
for municipio in glob.glob("./Jsons/*"):
    for rest in glob.glob(municipio + "/*.json"):
        with open(rest, "r", encoding="utf-8") as f:
            data = json.load(f)
            social_media_presence.append(data["Social Media"])

social_media_counts = {'Facebook': 0, 'Instagram': 0}
for presence in social_media_presence:
    for platform, exists in presence.items():
        if exists:
            social_media_counts[platform] += 1

# Crear gráfico de barras
plt.figure(figsize=(10, 6))
plt.bar(social_media_counts.keys(), social_media_counts.values())
plt.title('Presencia en Redes Sociales')
plt.xlabel('Red Social')
plt.ylabel('Número de Restaurantes')
plt.xticks(rotation=45)
plt.show()
```


Más allá de la comida, los servicios ofrecidos por los restaurantes pueden hacer que una experiencia sea memorable. Desde áreas para fumadores hasta servicios de entrega a domicilio, este análisis nos permite entender qué tan completos son los servicios de cada establecimiento.

Por ejemplo, algunos restaurantes ofrecen áreas privadas para eventos especiales, catering, y menús vegetarianos y veganos, lo que los hace ideales para una variedad de ocasiones y preferencias dietéticas. Otros pueden tener servicios adicionales como estacionamiento, áreas de juego para niños y entretenimiento en vivo. Así, puedes elegir el lugar que mejor se adapte a tus necesidades y preferencias, asegurando una experiencia gastronómica completa y satisfactoria

```python
# Análisis de servicios ofrecidos
services = []
for municipio in glob.glob("./Jsons/*"):
    for rest in glob.glob(municipio + "/*.json"):
        with open(rest, "r", encoding="utf-8") as f:
            data = json.load(f)
            services.append(data["Place Options"])

service_counts = {}
for service_set in services:
    for service, available in service_set.items():
        if available:
            if service in service_counts:
                service_counts[service] += 1
            else:
                service_counts[service] = 1

# Crear gráfico de barras
plt.figure(figsize=(10, 6))
plt.bar(service_counts.keys(), service_counts.values())
plt.title('Servicios Ofrecidos por Restaurantes')
plt.xlabel('Servicio')
plt.ylabel('Número de Restaurantes')
plt.xticks(rotation=45)
plt.show()
```
